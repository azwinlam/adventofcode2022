```python
with open("./day2data.txt", encoding = 'utf-8') as f:
   data = [line.strip() for line in f.readlines()]

array_data = data
array_data
```

```python
# A - ROCK +1 . - X
# B - Paper +2  - Y
# C - Scissor +3 - Z
```

```python
modified = [array.split(" ") for array in array_data]
total = 0
for i in modified:
    if i[0] == "A":
        if i[1] == "Y":
            i.append(2)
            i.append(6)
            total += 8
        if i[1] == "Z":
            i.append(3)
            i.append(0)
            total += 3
        if i[1] == "X":
            i.append(1)
            i.append(3)
            total += 4
    if i[0] == "B":
        if i[1] == "Z":
            i.append(3)
            i.append(6)
            total += 9
        if i[1] == "X":
            i.append(1)
            i.append(0)
            total += 1
        if i[1] == "Y":
            i.append(2)
            i.append(3)
            total += 5
    if i[0] == "C":
        if i[1] == "X":
            i.append(1)
            i.append(6)
            total += 7
        if i[1] == "Y":
            i.append(2)
            i.append(0)
            total += 2
        if i[1] == "Z":
            i.append(3)
            i.append(3)
            total += 6
total
```
